<h2>Which Loss Function to Use?</h2>

<h3>1. For Regression Tasks</h3>
<table border="1" cellpadding="5">
  <tr>
    <th>Loss Function</th>
    <th>When to Use</th>
    <th>Bengali Explanation</th>
  </tr>
  <tr>
    <td>Mean Squared Error (MSE)</td>
    <td>Default for regression (penalizes large errors)</td>
    <td>যখন বড় ভুলকে বেশি গুরুত্ব দিতে চাই</td>
  </tr>
  <tr>
    <td>Mean Absolute Error (MAE)</td>
    <td>When errors should be treated equally</td>
    <td>সব ভুলকে সমানভাবে গুরুত্ব দিলে</td>
  </tr>
  <tr>
    <td>Huber Loss</td>
    <td>When you want balance between MSE and MAE</td>
    <td>Outlier এর presence থাকলে</td>
  </tr>
  <tr>
    <td>Log-Cosh</td>
    <td>Smooth alternative to MSE</td>
    <td>MSE এর smoother রূপ</td>
  </tr>
</table>

<h3>2. For Binary Classification</h3>
<table border="1" cellpadding="5">
  <tr>
    <th>Loss Function</th>
    <th>When to Use</th>
    <th>Bengali Explanation</th>
  </tr>
  <tr>
    <td>Binary Cross-Entropy</td>
    <td>When output is 0 or 1</td>
    <td>দুটি class এর মধ্যে যেকোনো একটি নির্ধারণ করতে</td>
  </tr>
  <tr>
    <td>Hinge Loss</td>
    <td>When using Support Vector Machines (SVM)</td>
    <td>SVM এর ক্ষেত্রে ব্যবহৃত হয়</td>
  </tr>
</table>

<h3>3. For Multi-Class Classification</h3>
<table border="1" cellpadding="5">
  <tr>
    <th>Loss Function</th>
    <th>When to Use</th>
    <th>Bengali Explanation</th>
  </tr>
  <tr>
    <td>Categorical Cross-Entropy</td>
    <td>When labels are one-hot encoded</td>
    <td>One-hot encoded label থাকলে</td>
  </tr>
  <tr>
    <td>Sparse Categorical Cross-Entropy</td>
    <td>When labels are integer (0, 1, 2, ...)</td>
    <td>লেবেলগুলো integer হলে (one-hot না হলে)</td>
  </tr>
</table>
