  <h1>Vanishing Gradient এবং Sigmoid ফাংশনের সম্পর্ক</h1>

  <h2>Vanishing Gradient সমস্যা কী?</h2>
  <p>
    Vanishing Gradient হলো এমন একটি সমস্যা যেখানে ব্যাকপ্রপাগেশনের সময় ডেরিভেটিভগুলো এত ছোট হয়ে যায় যে,
    ওজন (weight) ঠিকভাবে আপডেট হয় না। এর ফলে নিউরাল নেটওয়ার্ক ভালোভাবে শেখে না, বিশেষ করে যখন লেয়ারের সংখ্যা অনেক বেশি হয়।
  </p>

  <h2>Sigmoid ফাংশন ও এর ডেরিভেটিভ</h2>
  <div class="section">
    σ(x) = 1 / (1 + e<sup>-x</sup>)<br>
    σ′(x) = σ(x) × (1 - σ(x))
  </div>

  <p>
    যখন ইনপুট খুব বড় ধনাত্মক বা ঋণাত্মক হয়, তখন Sigmoid ফাংশনের আউটপুট হয় প্রায় ১ বা ০। তখন তার ডেরিভেটিভ হয়ে পড়ে প্রায় শূন্য,
    যার ফলে gradient vanish করে অর্থাৎ ধীরে ধীরে ছোট হতে হতে হারিয়ে যায়।
  </p>

  <h2>ডেরিভেটিভের মানের উদাহরণ</h2>

  <h2>সমস্যার প্রভাব</h2>
  <ul>
    <li>Gradient খুব ছোট হয়ে যায়।</li>
    <li>ওজন আপডেট হয় না বা হয় খুব ধীরে।</li>
    <li>Deep Layer-এ শেখা বন্ধ হয়ে যায়।</li>
  </ul>

  <h2>সমাধান</h2>
  <p>Sigmoid-এর পরিবর্তে নিচের অ্যাক্টিভেশন ফাংশনগুলো ব্যবহার করলে Vanishing Gradient সমস্যা অনেকাংশে এড়ানো যায়:</p>
