<h2>Min, Max, Average Pooling in CNN</h2>

<h3>Pooling কী?</h3>
<p>
Pooling হলো CNN-এ ব্যবহৃত একটি অপারেশন যা ফিচার ম্যাপের সাইজ কমাতে সাহায্য করে। এটি ফিচারের <strong>dimensionality কমায়</strong> এবং <strong>computational cost হ্রাস করে</strong>। Pooling অপারেশন ফিচার ম্যাপের উপর একটি ছোট উইন্ডো চালিয়ে প্রতি অঞ্চলের একটি ভ্যালু রাখে।
</p>

<h3>১. Max Pooling</h3>
<p>
<strong>Max Pooling</strong> প্রতিটি region/window-র মধ্যে সর্বোচ্চ মানটি নির্বাচন করে। এটি সবচেয়ে গুরুত্বপূর্ণ বা শক্তিশালী ফিচারটিকে ধরে রাখে।
</p>

<pre>
Input Patch:
[[2, 4],
 [3, 1]]

Max Pooling Result: 4
</pre>

<h4>বৈশিষ্ট্য:</h4>
<ul>
  <li>সবচেয়ে জনপ্রিয় পুলিং</li>
  <li>শক্তিশালী ফিচার সংরক্ষণ করে</li>
  <li>Noise কমায়</li>
</ul>

<h3>২. Average Pooling</h3>
<p>
<strong>Average Pooling</strong> প্রতিটি window-র সব মানের গড় বের করে। এটি একটু বেশি smooth result দেয়।
</p>

<pre>
Input Patch:
[[2, 4],
 [3, 1]]

Average Pooling Result: (2+4+3+1)/4 = 2.5
</pre>

<h4>বৈশিষ্ট্য:</h4>
<ul>
  <li>ফিচার গড় মান নেয়</li>
  <li>Smooth feature map তৈরি করে</li>
  <li>Less aggressive than max pooling</li>
</ul>

<h3>৩. Min Pooling</h3>
<p>
<strong>Min Pooling</strong> প্রতিটি window থেকে সর্বনিম্ন মান নির্বাচন করে। এটি খুব কম ব্যবহৃত হয়।
</p>

<pre>
Input Patch:
[[2, 4],
 [3, 1]]

Min Pooling Result: 1
</pre>

<h4>বৈশিষ্ট্য:</h4>
<ul>
  <li>সবচেয়ে ছোট মান নেয়</li>
  <li>Noise-sensitive</li>
  <li>কম ব্যবহার হয় বাস্তব সমস্যায়</li>
</ul>

<h3>উপসংহার</h3>
<p>
Pooling অপারেশন CNN-এ গুরুত্বপূর্ণ ভূমিকা পালন করে। Max pooling সবচেয়ে কার্যকর এবং প্রচলিত, কারণ এটি সবচেয়ে গুরুত্বপূর্ণ ফিচার ধরে রাখে। তবে প্রয়োজনে Average বা Min pooling-ও ব্যবহার করা যেতে পারে।
</p>
