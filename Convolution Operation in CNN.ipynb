<h2>Convolution Operation in CNN</h2>

<h3>কনভোলিউশন অপারেশন কী?</h3>
<p>
কনভোলিউশন অপারেশন হলো Convolutional Neural Network (CNN)-এর প্রধান কার্যক্রম, যেখানে একটি ছোট ম্যাট্রিক্স (যাকে <strong>ফিল্টার</strong> বা <strong>কর্ণেল</strong> বলা হয়) ইনপুট ইমেজের উপর ঘোরানো হয় (slide করা হয়) এবং প্রতিটি অঞ্চলের সাথে গাণিতিকভাবে গুণ করে একটি নতুন আউটপুট তৈরি করা হয়। এটি মূলত ফিচার এক্সট্র্যাকশন এর জন্য ব্যবহৃত হয়।
</p>

<h3>মূল উপাদান:</h3>
<ul>
  <li><strong>Input Image:</strong> একটি ম্যাট্রিক্স বা টেনসর যা প্রতিটি পিক্সেল ভ্যালু দ্বারা গঠিত।</li>
  <li><strong>Kernel/Filter:</strong> ছোট আকারের একটি ম্যাট্রিক্স (যেমন 3x3), যা ইমেজের উপর দিয়ে স্লাইড করে।</li>
  <li><strong>Stride:</strong> ফিল্টার কত ধাপে ইনপুট ইমেজের উপর স্লাইড করবে।</li>
  <li><strong>Padding:</strong> ইমেজের কিনারায় অতিরিক্ত পিক্সেল যোগ করে যাতে আউটপুট সাইজ ছোট না হয়।</li>
</ul>

<h3>কাজের ধাপ:</h3>
<ol>
  <li>ফিল্টার ইনপুট ইমেজের এক অংশের উপর রাখা হয়।</li>
  <li>সেই অংশের প্রতিটি পিক্সেলকে ফিল্টারের সংশ্লিষ্ট পজিশনের পিক্সেলের সাথে গুণ করা হয়।</li>
  <li>সবগুলো গুণফলের যোগফল নেয়া হয়।</li>
  <li>এই মানটি আউটপুট ফিচার ম্যাপে রাখা হয়।</li>
  <li>এরপর ফিল্টারটি স্লাইড করে পরবর্তী অঞ্চলে নিয়ে যাওয়া হয়।</li>
</ol>

<h3>উদাহরণ (3x3 ফিল্টার):</h3>

<pre>
Input Image:
[[1, 2, 0],
 [4, 5, 6],
 [7, 8, 9]]

Filter:
[[1, 0, -1],
 [1, 0, -1],
 [1, 0, -1]]

Convolution Result (without padding):
= 1*1 + 2*0 + 0*(-1) +
  4*1 + 5*0 + 6*(-1) +
  7*1 + 8*0 + 9*(-1)
= 1 + 0 + 0 + 4 + 0 - 6 + 7 + 0 - 9
= -3
</pre>

<h3>Output Feature Map:</h3>
<p>
ফিল্টারটি পুরো ইমেজ জুড়ে চালানোর পরে যে ম্যাট্রিক্সটি তৈরি হয়, তাকে Feature Map বা Activation Map বলা হয়।
</p>

<h3>উদ্দেশ্য:</h3>
<ul>
  <li>ইমেজ থেকে গুরুত্বপূর্ণ প্যাটার্ন (যেমন edges, textures, shapes) এক্সট্র্যাক্ট করা।</li>
  <li>নতুন ফিচার তৈরির মাধ্যমে মডেলকে ক্লাসিফিকেশনে সাহায্য করা।</li>
</ul>

<h3>উপসংহার:</h3>
<p>
Convolution operation CNN-এ ইমেজ প্রসেসিং ও ফিচার এক্সট্র্যাকশনের মূলে রয়েছে। এটি ইনপুট ইমেজের গুরুত্বপূর্ণ বৈশিষ্ট্যগুলোকে হাইলাইট করে এবং ক্লাসিফিকেশন বা ডিটেকশনে সহায়তা করে।
</p>
