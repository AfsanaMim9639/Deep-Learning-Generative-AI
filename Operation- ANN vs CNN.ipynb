<h2>ANN vs CNN - অপারেশনের তুলনা</h2>

<h3>১. Artificial Neural Network (ANN) কীভাবে কাজ করে?</h3>
<p>
ANN একটি সাধারণ নিউরাল নেটওয়ার্ক, যেখানে প্রতিটি লেয়ার হলো <strong>fully connected</strong> অর্থাৎ প্রতিটি নিউরন পরবর্তী লেয়ারের প্রতিটি নিউরনের সাথে সংযুক্ত। এটি সাধারণভাবে ট্যাবুলার ডেটার জন্য ব্যবহৃত হয়।
</p>

<ul>
  <li><strong>Input:</strong> 1D vector (যেমন: [x₁, x₂, x₃, ...])</li>
  <li><strong>Each Layer:</strong> Matrix multiplication (Weight × Input + Bias)</li>
  <li><strong>Activation Function:</strong> ReLU, Sigmoid, tanh ইত্যাদি</li>
  <li><strong>Drawback:</strong> ইমেজের spatial structure ধরে রাখতে পারে না</li>
</ul>

<h3>২. Convolutional Neural Network (CNN) কীভাবে কাজ করে?</h3>
<p>
CNN হলো এক ধরনের deep neural network যা ইমেজ বা স্প্যাটিয়াল ডেটা প্রসেস করতে ব্যবহৃত হয়। এটি <strong>convolution operation</strong> দিয়ে ফিচার এক্সট্র্যাক্ট করে এবং তারপর classification বা detection করে।
</p>

<ul>
  <li><strong>Input:</strong> 2D (image) or 3D (RGB image)</li>
  <li><strong>Layers:</strong> Convolution → Activation → Pooling → Fully Connected</li>
  <li><strong>Main Operation:</strong> Filters/Kernel দিয়ে sliding করে local features বের করা</li>
  <li><strong>Advantage:</strong> Edge, shape, texture detect করতে পারে</li>
</ul>

<h3>৩. তুলনামূলক টেবিল</h3>
<table border="1" cellpadding="5">
  <tr>
    <th>বিষয়</th>
    <th>ANN</th>
    <th>CNN</th>
  </tr>
  <tr>
    <td>Input Type</td>
    <td>1D vector</td>
    <td>2D/3D image</td>
  </tr>
  <tr>
    <td>Connectivity</td>
    <td>Fully connected neurons</td>
    <td>Locally connected via filters</td>
  </tr>
  <tr>
    <td>Main Operation</td>
    <td>Matrix multiplication (Weight × Input)</td>
    <td>Convolution operation (Filter × Patch)</td>
  </tr>
  <tr>
    <td>Parameters</td>
    <td>অনেক বেশি (more prone to overfitting)</td>
    <td>কম (shared weights reduce parameters)</td>
  </tr>
  <tr>
    <td>Feature Extraction</td>
    <td>Manually needed</td>
    <td>Automatically through convolution</td>
  </tr>
  <tr>
    <td>Use Case</td>
    <td>Tabular data, structured input</td>
    <td>Image, video, spatial data</td>
  </tr>
</table>

<h3>৪. উপসংহার</h3>
<p>
ANN মূলত ফিচার ফ্ল্যাট করে ইনপুট নেয়, যেখানে CNN ইমেজের স্থানিক (spatial) গঠন ধরে রাখে এবং layeredভাবে ফিচার এক্সট্র্যাক্ট করে। ইমেজ বা ভিশন ভিত্তিক সমস্যার জন্য CNN অধিক কার্যকর।
</p>
