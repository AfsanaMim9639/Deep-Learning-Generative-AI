<h2>ছবি (Image) সম্পর্কে সবকিছু</h2>

<h3>১. ইমেজ কী?</h3>
<p>
ইমেজ হলো একধরনের ভিজ্যুয়াল ডেটা যা পিক্সেল (pixel) দ্বারা গঠিত। প্রতিটি পিক্সেল একটি রঙ বা intensity মান বহন করে। ইমেজ কম্পিউটারের জন্য একটি মাল্টি-ডাইমেনশনাল অ্যারে (matrix) হিসাবে কাজ করে।
</p>

<h3>২. ইমেজ টাইপ (Types of Images)</h3>
<ul>
  <li><strong>Grayscale Image:</strong> কেবলমাত্র intensity থাকে (black & white)। প্রতিটি পিক্সেল ০ থেকে ২৫৫ মানের মধ্যে হয়।</li>
  <li><strong>RGB Image:</strong> প্রতিটি পিক্সেলের জন্য তিনটি চ্যানেল (Red, Green, Blue)।</li>
  <li><strong>RGBA Image:</strong> RGB + Alpha (Transparency) চ্যানেল।</li>
  <li><strong>Binary Image:</strong> শুধু ০ (black) এবং ১ (white)।</li>
</ul>

<h3>৩. ইমেজের ডাইমেনশন</h3>
<ul>
  <li>Grayscale: (Height x Width)</li>
  <li>RGB: (Height x Width x 3)</li>
  <li>RGBA: (Height x Width x 4)</li>
</ul>

<h3>৪. ইমেজ রিডিং ও প্রসেসিং (Python Example)</h3>
<pre>
import cv2
import matplotlib.pyplot as plt

# ইমেজ লোড করা
img = cv2.imread('image.jpg')  # BGR format
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# ইমেজ দেখানো
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
</pre>

<h3>৫. ইমেজ প্রসেসিং (Image Processing)</h3>

  <li>Crop</li>
  <li>Rotate</li>
  <li>Filter (Blur, Sharpen)</li>
  <li>Edge Detection (Canny, Sobel)</li>
</ul>

<h3>৬. ইমেজ ফাইল ফরম্যাট</h3>
<ul>
  <li><strong>JPEG/JPG:</strong> Compressed, lossy</li>
  <li><strong>PNG:</strong> Lossless, supports transparency</li>
  <li><strong>BMP:</strong> Uncompressed, large size</li>
  <li><strong>TIFF:</strong> High-quality image, used in medical and printing</li>
</ul>

<h3>৭. ইমেজ ডেটা প্রিপ্রসেসিং (Model Input এর জন্য)</h3>
<ul>
  <li>Normalization (পিক্সেল মান ০ থেকে ১ এ স্কেল করা)</li>
  <li>Resizing (উদাহরণ: 224x224 for CNN)</li>
  <li>Augmentation (Rotate, Flip, Zoom)</li>
</ul>

<h3>৯. ইমেজ ও Deep Learning</h3>
<p>
Deep Learning-এ ইমেজ প্রক্রিয়াকরণের জন্য Convolutional Neural Networks (CNN) সবচেয়ে বেশি ব্যবহৃত হয়। এটি ইমেজ থেকে ফিচার এক্সট্র্যাক্ট করে ক্লাসিফিকেশন বা সেগমেন্টেশনের জন্য প্রস্তুত করে।
</p>
