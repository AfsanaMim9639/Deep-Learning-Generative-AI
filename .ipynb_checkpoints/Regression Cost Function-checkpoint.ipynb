<h2>Regression Cost Functions</h2>

<h3>1. Mean Squared Error (MSE)</h3>
<ul>
  <li><b>Formula:</b> MSE = (1/n) * Σ (y<sub>i</sub> - ŷ<sub>i</sub>)²</li>
  <li><b>English:</b> Calculates the average of the squared differences between actual and predicted values.</li>
  <li><b>বাংলা:</b> আসল এবং পূর্বানুমান করা মানগুলোর পার্থকের বর্গের গড় বের করে।</li>
  <li><b>Usage:</b> Most common for regression tasks.</li>
</ul>

<h3>2. Mean Absolute Error (MAE)</h3>
<ul>
  <li><b>Formula:</b> MAE = (1/n) * Σ |y<sub>i</sub> - ŷ<sub>i</sub>|</li>
  <li><b>English:</b> Calculates the average of the absolute differences between actual and predicted values.</li>
  <li><b>বাংলা:</b> আসল এবং পূর্বানুমান মানের মধ্যকার পার্থকের পরম মানের গড়।</li>
  <li><b>Usage:</b> When outliers should not be squared (less sensitive to outliers).</li>
</ul>

<h3>3. Huber Loss</h3>
<ul>
  <li><b>Formula:</b> Combines MSE and MAE. It's quadratic when error is small and linear when error is large.</li>
  <li><b>English:</b> Less sensitive to outliers than MSE, more stable than MAE.</li>
  <li><b>বাংলা:</b> ছোট error এর জন্য squared error ব্যবহার করে এবং বড় error এর জন্য absolute error।</li>
  <li><b>Usage:</b> Balanced choice when you want to be robust to outliers.</li>
</ul>

<h3>4. Log-Cosh Loss</h3>
<ul>
  <li><b>Formula:</b> log(cosh(ŷ<sub>i</sub> - y<sub>i</sub>))</li>
  <li><b>English:</b> Similar to MSE but less sensitive to large errors.</li>
  <li><b>বাংলা:</b> MSE এর মত কাজ করে, তবে বড় error এর প্রতি কম সংবেদনশীল।</li>
  <li><b>Usage:</b> Smooth alternative to MSE and MAE.</li>
</ul>
