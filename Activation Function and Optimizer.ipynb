<h2>Activation Functions vs Optimizers</h2>

<h3>1. Activation Functions</h3>

<h4>What is an Activation Function?</h4>
<ul>
  <li><b>English:</b> An activation function determines the output of a neural network node (neuron) based on its input. It introduces non-linearity into the model, allowing it to learn more complex patterns.</li>
  <li><b>বাংলা:</b> Activation function হল একটি ফাংশন যা নেটওয়ার্কের প্রতিটি নিউরনের ইনপুটের উপর ভিত্তি করে আউটপুট নির্ধারণ করে। এটি মডেলে নন-লিনিয়ারিটি যুক্ত করে, যা মডেলকে আরও জটিল প্যাটার্ন শেখাতে সাহায্য করে।</li>
</ul>

<h4>When to Use Activation Functions?</h4>
<ul>
  <li><b>English:</b> You use activation functions in each node of a neural network, typically after the weighted sum (linear combination) of inputs.</li>
  <li><b>বাংলা:</b> নিউরাল নেটওয়ার্কের প্রতিটি নোডে ইনপুটের ওজনযুক্ত যোগফলের পর activation function ব্যবহার করা হয়।</li>
</ul>

<h4>Common Activation Functions</h4>
<ul>
  <li><b>Sigmoid:</b> Used for binary classification problems.</li>
  <li><b>ReLU:</b> Most common in hidden layers for deep learning models.</li>
  <li><b>Tanh:</b> Often used when values need to be between -1 and 1.</li>
  <li><b>Softmax:</b> Used in the output layer for multi-class classification.</li>
</ul>

<h3>2. Optimizers</h3>

<h4>What is an Optimizer?</h4>
<ul>
  <li><b>English:</b> An optimizer is an algorithm used to minimize (or maximize) a loss function by adjusting the weights of the model. The optimizer controls how the weights are updated during training.</li>
  <li><b>বাংলা:</b> Optimizer একটি এলগরিদম যা লস ফাংশনকে সর্বনিম্ন (বা সর্বোচ্চ) করতে মডেলের ওয়েটস আপডেট করে। এটি ট্রেনিংয়ের সময় ওয়েটস আপডেট করার পদ্ধতি নিয়ন্ত্রণ করে।</li>
</ul>

<h4>When to Use Optimizers?</h4>
<ul>
  <li><b>English:</b> Optimizers are used during the training phase of a neural network model. They are responsible for adjusting the model’s weights and bias to minimize the loss function.</li>
  <li><b>বাংলা:</b> Optimizer ব্যবহার করা হয় মডেল ট্রেনিং এর সময়। এটি মডেলের ওয়েট এবং বায়াস আপডেট করে লস ফাংশন মিনিমাইজ করতে।</li>
</ul>

<h4>Common Optimizers</h4>
<ul>
  <li><b>SGD (Stochastic Gradient Descent):</b> Basic optimizer, uses the gradient of the loss function to update weights.</li>
  <li><b>Adam:</b> Adaptive moment estimation, commonly used due to its efficiency and fast convergence.</li>
  <li><b>RMSProp:</b> Optimizer that adapts the learning rate based on recent gradients.</li>
</ul>

<h3>Summary</h3>
<ul>
  <li><b>English:</b> Activation functions are used to add non-linearity to the neural network and control its output, while optimizers are used to train the model by adjusting the weights to minimize the loss function.</li>
  <li><b>বাংলা:</b> Activation function নিউরাল নেটওয়ার্কে নন-লিনিয়ারিটি যোগ করতে এবং আউটপুট নিয়ন্ত্রণ করতে ব্যবহৃত হয়, আর Optimizer মডেলকে ট্রেন করতে এবং ওয়েট আপডেট করে লস ফাংশন মিনিমাইজ করতে ব্যবহৃত হয়।</li>
</ul>
