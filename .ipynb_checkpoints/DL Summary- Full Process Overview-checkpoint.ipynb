<h2>Deep Learning Summary (All Process Overview)</h2>

<h3>১. Deep Learning-এর ধরণ (Types of DL Models)</h3>
<ul>
  <li><strong>ANN (Artificial Neural Network):</strong> Structured data-র জন্য</li>
  <li><strong>CNN (Convolutional Neural Network):</strong> Image/Video data</li>
  <li><strong>RNN (Recurrent Neural Network):</strong> Sequence data (Text, Time Series)</li>
  <li><strong>GAN (Generative Adversarial Network):</strong> Image generation</li>
  <li><strong>Autoencoder:</strong> Dimensionality reduction বা anomaly detection</li>
</ul>

<hr/>

<h3>২. Model Architecture Layers</h3>
<ul>
  <li><strong>Input Layer:</strong> ডেটা প্রবেশ করে</li>
  <li><strong>Hidden Layers:</strong> শিখে ফিচার, Conv বা Dense</li>
  <li><strong>Output Layer:</strong> Prediction দেয়</li>
</ul>

<h4>Special Layers:</h4>
<ul>
  <li><strong>Flatten Layer:</strong> Conv এর পর 1D vector বানায়</li>
  <li><strong>Dropout Layer:</strong> Overfitting কমাতে কিছু neuron বন্ধ করে দেয়</li>
  <li><strong>Batch Normalization:</strong> Training Stability বাড়ায়</li>
</ul>

<hr/>

<h3>৩. Activation Functions</h3>
<ul>
  <li><strong>ReLU:</strong> Hidden layer-এ সবচেয়ে বেশি ব্যবহৃত</li>
  <li><strong>Sigmoid:</strong> Binary classification-এর output layer</li>
  <li><strong>tanh:</strong> Centered output, RNN-এ ব্যবহৃত</li>
  <li><strong>Softmax:</strong> Multiclass classification</li>
  <li><strong>Leaky ReLU / PReLU:</strong> ReLU-এর উন্নত সংস্করণ (negative অংশের সমস্যা সমাধান)</li>
</ul>

<hr/>

<h3>৪. Output Layer & Loss Function</h3>
<table border="1" cellpadding="5">
  <tr>
    <th>Problem Type</th>
    <th>Output Layer</th>
    <th>Loss Function</th>
  </tr>
  <tr>
    <td>Binary Classification</td>
    <td>1 unit + Sigmoid</td>
    <td>Binary Crossentropy</td>
  </tr>
  <tr>
    <td>Multiclass Classification</td>
    <td>Softmax</td>
    <td>Categorical Crossentropy</td>
  </tr>
  <tr>
    <td>Regression</td>
    <td>Linear</td>
    <td>MSE / MAE</td>
  </tr>
</table>

<hr/>

<h3>৫. Optimizers (Gradient Descent based)</h3>
<ul>
  <li><strong>SGD:</strong> সাধারণ গ্রেডিয়েন্ট ডিসেন্ট</li>
  <li><strong>SGD with Momentum:</strong> গতির সাথে optimize করে</li>
  <li><strong>RMSProp:</strong> Learning rate adaptively adjust করে</li>
  <li><strong>Adam:</strong> সবচেয়ে জনপ্রিয়, adaptive + momentum</li>
</ul>

<hr/>

<h3>৬. Regularization Techniques</h3>
<ul>
  <li><strong>Dropout:</strong> কিছু neuron randomly বন্ধ করে</li>
  <li><strong>L1/L2 Regularization:</strong> Weight কম রাখে</li>
  <li><strong>Early Stopping:</strong> Validation loss না কমলে training বন্ধ</li>
</ul>

<hr/>

<h3>৭. Training Process Overview</h3>
<ol>
  <li>ডেটা প্রিপ্রসেসিং (scaling, reshaping)</li>
  <li>Model architecture define (layer, activation)</li>
  <li>Compile (optimizer, loss, metrics)</li>
  <li>Train (epochs, batch_size)</li>
  <li>Evaluate/Test</li>
  <li>Prediction</li>
</ol>

<hr/>

<h3>৮. Evaluation Metrics</h3>
<ul>
  <li><strong>Accuracy:</strong> Classification</li>
  <li><strong>Precision, Recall, F1-score:</strong> Imbalanced data</li>
  <li><strong>RMSE, MAE:</strong> Regression</li>
</ul>


<h2>Neural Network Example with Activation Functions & Optimizer (without Icons)</h2>

<h3>১. Activation Function কেন দরকার?</h3>
<p>
Activation function একটি গুরুত্বপূর্ণ উপাদান যা নিউরাল নেটওয়ার্কে non-linearity যোগ করে, যাতে মডেল জটিল সম্পর্ক শিখতে পারে।
</p>

<h3>২. Activation Function কোথায় ব্যবহার করবো?</h3>
<p>
- Hidden Layers-এ Activation Function ব্যবহার হয়।<br/>
- Output Layer-এ Activation Function নির্ভর করে সমস্যার ধরণে:
  <ul>
    <li>Binary Classification: Sigmoid</li>
    <li>Multiclass Classification: Softmax</li>
    <li>Regression: Linear</li>
  </ul>
</p>

<h3>৩. Activation Functions এর ধরন:</h3>
<ul>
  <li><strong>ReLU:</strong> Hidden layers-এ খুব জনপ্রিয়, Negative values কে 0 করে দেয়</li>
  <li><strong>Sigmoid:</strong> Binary classification-এর জন্য, Output value 0 থেকে 1 এর মধ্যে থাকে</li>
  <li><strong>tanh:</strong> Centered output (−1 to 1), RNN-এ বেশি ব্যবহৃত</li>
  <li><strong>Softmax:</strong> Multiclass classification-এ, class probabilities বের করে</li>
  <li><strong>Linear:</strong> Regression problems-এ ব্যবহার হয়, real number output দেয়</li>
</ul>

<h3>৪. Optimizer কী এবং কবে ব্যবহার করবো?</h3>
<p>
Optimizer হল এমন একটি এলগরিদম যা নিউরাল নেটওয়ার্কের ওজন (weights) আপডেট করে যাতে <strong>loss function</strong> মিনিমাইজ হয় এবং মডেল ধীরে ধীরে ভাল ফলাফল দিতে শেখে।
</p>

<h4>কখন ব্যবহার করবো?</h4>
<ul>
  <li>প্রতিটি মডেল compile করার সময় optimizer প্রয়োজন হয়</li>
  <li>Optimizer ছাড়া Gradient Descent কাজ করতে পারে না</li>
</ul>

<h4>প্রধান Optimizer গুলো:</h4>
<ul>
  <li><strong>SGD (Stochastic Gradient Descent):</strong> সোজাসাপটা, ধীরে converge করে</li>
  <li><strong>SGD with Momentum:</strong> আগে যেদিকে গিয়েছিল সেই দিকের ইনফ্লুয়েন্স রাখে</li>
  <li><strong>RMSProp:</strong> Learning rate auto-adjust করে, time-series data-তে ভালো</li>
  <li><strong>Adam:</strong> সবচেয়ে জনপ্রিয়; RMSProp + Momentum এর সংমিশ্রণ</li>
</ul>

<p><strong>বেশিরভাগ সময় আমরা Adam ব্যবহার করি কারণ এটি দ্রুত এবং নির্ভরযোগ্য।</strong></p>

<h3>৫. Neural Network-এর পূর্ণ প্রসেস:</h3>
<ol>
  <li>Input Layer: ইনপুট ডেটা গ্রহণ</li>
  <li>Hidden Layer: Dense + Activation Function (ReLU)</li>
  <li>Output Layer: Sigmoid বা অন্য Activation Function</li>
  <li>Compile Stage: Optimizer + Loss Function নির্বাচন</li>
  <li>Training Stage: Weight আপডেট করা হয় optimizer দিয়ে</li>
</ol>

<h3>৬. Neural Network Example (Binary Classification)</h3>
<pre>
Input (x1, x2, x3) → Dense Layer (4 neuron) → ReLU → Dense Layer (1 neuron) → Sigmoid → Output (0 or 1)
</pre>

<h3>৭. Keras Code Example:</h3>
<pre>
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Build model
model = Sequential()

# Input layer + Hidden layer
model.add(Dense(4, input_dim=3, activation='relu'))  # input_dim=3 because x1, x2, x3

# Output layer
model.add(Dense(1, activation='sigmoid'))  # Binary classification → sigmoid

# Compile model with optimizer
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
# model.fit(X_train, y_train, epochs=20, batch_size=16)
</pre>

<h3>৮. উপসংহার:</h3>
<p>
Activation Function নিউরাল নেটওয়ার্কে non-linearity যোগ করে, আর Optimizer ওজন আপডেট করে যাতে মডেল ধীরে ধীরে ভালো ফলাফল দিতে পারে। এ দুটি অংশ মডেলের সঠিক কাজের জন্য অপরিহার্য।
</p>
