<h2>Loss Functions for Classification</h2>

<h3>1. Binary Cross-Entropy (Log Loss)</h3>
<ul>
  <li><b>Formula:</b> 
    <br>− [ y log(ŷ) + (1−y) log(1−ŷ) ]
  </li>
  <li><b>English:</b> Measures the performance of a classification model whose output is a probability between 0 and 1.</li>
  <li><b>বাংলা:</b> যখন model দুইটি class (0 বা 1) এর মধ্যে বাছাই করে, তখন এটি actual ও predicted probability এর পার্থক্য হিসাব করে।</li>
  <li><b>Used for:</b> Binary classification (e.g., spam detection)</li>
</ul>

<h3>2. Categorical Cross-Entropy</h3>
<ul>
  <li><b>Formula:</b> 
    <br>− Σ y<sub>i</sub> log(ŷ<sub>i</sub>)
  </li>
  <li><b>English:</b> Compares the predicted probability distribution across multiple classes with the actual class (one-hot encoded).</li>
  <li><b>বাংলা:</b> একাধিক শ্রেণির মধ্যে একটি সঠিক class নির্বাচনের জন্য ব্যবহৃত হয়।</li>
  <li><b>Used for:</b> Multi-class classification (e.g., image classification)</li>
</ul>

<h3>3. Sparse Categorical Cross-Entropy</h3>
<ul>
  <li><b>English:</b> Same as categorical cross-entropy, but works when target labels are integers instead of one-hot vectors.</li>
  <li><b>বাংলা:</b> Categorical cross-entropy এর মতই, তবে এখানে label গুলো one-hot encoding এর বদলে integer আকারে থাকে (যেমন 0, 1, 2)।</li>
  <li><b>Used for:</b> Multi-class classification with integer labels</li>
</ul>

<h3>4. Hinge Loss</h3>
<ul>
  <li><b>Formula:</b> max(0, 1 − y × ŷ)</li>
  <li><b>English:</b> Commonly used for Support Vector Machines (SVMs). Encourages correct classification with margin.</li>
  <li><b>বাংলা:</b> SVM মডেলগুলোতে ব্যবহৃত হয়, যেখানে সঠিকভাবে class আলাদা করতে হলে নির্দিষ্ট একটি মার্জিন এর প্রয়োজন হয়।</li>
  <li><b>Used for:</b> Binary classification with SVM</li>
</ul>
